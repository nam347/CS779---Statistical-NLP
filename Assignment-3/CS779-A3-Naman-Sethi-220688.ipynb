## **1. Introduction to Word2Vec**

### **What is Word2Vec?**
Word2Vec is a shallow neural network that learns **dense vector representations of words** such that semantically similar words lie close together in the embedding space.

The key idea: instead of representing words as one-hot vectors, we represent them as low-dimensional **embeddings** learned by predicting context words from a target word (or vice versa).

---
### Skip-gram Architecture

The **Skip-gram model** in Word2Vec learns word embeddings by predicting context words given a center (target) word. Instead of using one-hot vectors, we use Label Encoding, which is more memory-efficient and computationally optimal.

####  Intuition

#### Architecture and Training Flow (with Index-based Input)

Let:  
- \( V \): size of the vocabulary  
- \( d \): embedding dimension  
-  t in $ 0, 1, \dots, V-1 $: index of the target word  
-  $E \in {R}^{V \times d}$: input embedding matrix  
-  $E' \in {R}^{V \times d}$: output embedding matrix

The integer index \( t \) is used to perform a direct embedding lookup in matrix \( E \), avoiding one-hot vector multiplication.





---

####  Step-by-Step Computation:

1. **Input**: Integer ID of target word \( t \)

2. **Embedding Lookup** (instead of one-hot):

$$
u = E[t] \in \mathbb{R}^d
$$

3. **Score for Context Word \( c \)**:

$$
\text{score}(c) = u^\top E'[c]
$$

4. **Softmax over Vocabulary**:

$$
P(w_c \mid w_t) = \frac{\exp(u^\top E'[c])}{\sum_{w=1}^{V} \exp(u^\top E'[w])}
$$

---



#### Objective Function

Given a **center word** $ w_t $ , the Skip-gram model aims to **predict surrounding context words** within a fixed window size.


$$
{L}_{\text{skip-gram}} = \sum_{-k \le j \le k,\ j \ne 0} \log P(w_{t+j} \mid w_t)
$$


where


| Term &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;| Description |
|------|-------------|
| $ {L}_{{skip-gram}} $ | The **log-likelihood** objective for predicting the context words surrounding the center word $ w_t $ This is the quantity we aim to **maximize** during training. |
| $ k $| The **window size**, which defines how many words before and after the center word are considered as context. |
| $ j $ | The **offset** from the center word $ t $, ranging from $ -k $ to $ +k $, excluding $ j = 0 $ (which would be the center word itself). |
| $ w_t $ | The **center word** (target word) whose surrounding words we are trying to predict. |
| $ w_{t+j} $ | A **context word** at position $ t+j $ relative to the center word $ w_t $ |
| $ P(w_{t+j} \mid w_t) $ | The **conditional probability** that word $ w_{t+j} $ appears in the context of $ w_t $. This is typically modeled using a softmax function or approximated using negative sampling. |
| $ \log P(w_{t+j} \mid w_t) $ | The **log-probability** of correctly predicting a context word. Using the log helps with numerical stability and converts the product of probabilities into a sum. |



---




#### Parameters Learned

- $ E \in {R}^{V \times d} $: Embedding matrix (input)
- $ E' \in {R}^{V \times d} $: Output matrix (context prediction)

**Embedding Lookup** via integer index replaces costly one-hot vector multiplication.

**Note**: Final embeddings are typically taken from matrix $ E $
---


#### Summary

- **Input**: Integer index of target word  
- **Output**: Predict context word indices  
- **Lookup**: Directly fetch embedding vector from matrix  
- **Train**: Using softmax or negative sampling  
- **Goal**: Words appearing in similar contexts should have similar vectors

### CBOW Architecture

The **CBOW model** in Word2Vec learns word embeddings by predicting a center (target) word given its surrounding context words. Like Skip-gram, it avoids one-hot vectors using index-based embedding lookup, making it both space- and time-efficient.

#### Intuition

Sentence:  
**"The cat sat on the mat"**, with **window size = 2**

If target word = `"cat"`, context = `["The", "sat"]`  
→ Training pairs:
- (`["The", "sat"]` → `cat`)

#### Architecture and Training Flow (with Index-based Input)

Let:  
- $ V $: size of the vocabulary  
- $ d $: embedding dimension  
-   C = $ c_1, c_2, \dots, c_m $ : list of context word indices  
-  $ t $: index of the target word  
-  $ E \in {R}^{V \times d} $ : input embedding matrix  
-  $ E' \in {R}^{V \times d} $ : output embedding matrix

Each context word $ c_i \in C $ is used to lookup a vector from $ E $, which are then averaged to form a single context embedding.

---

#### Step-by-Step Computation:

1. **Input**: Integer IDs of context words  C = $ {c_1, c_2, \dots, c_m} $

2. **Embedding Lookup**:

$$
h = \frac{1}{m} \sum_{i=1}^{m} E[c_i] \in \mathbb{R}^d
$$

3. **Score for Target Word $ t $**:

$$
\text{score}(t) = h^\top E'[t]
$$

4. **Softmax over Vocabulary**:

$$
P(w_t \mid \text{context}) = \frac{\exp(h^\top E'[t])}{\sum_{w=1}^{V} \exp(h^\top E'[w])}
$$

---

#### Objective Function

Given **context words** $ \{w_{t-k}, \dots, w_{t-1}, w_{t+1}, \dots, w_{t+k}\} $, the CBOW model aims to **predict the center word** \( w_t \).

$$
{L}_{\text{CBOW}} = \log P(w_t \mid w_{t-k}, \dots, w_{t-1}, w_{t+1}, \dots, w_{t+k})
$$

where


| Term &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;    | Description |
|---------|-------------|
| $ {L}_{\text{CBOW}} $ | The **log-likelihood** objective for predicting the center word $ w_t $ given its surrounding context words. |
| $ k $ | The **window size**, which defines how many words before and after the target word are used as context. |
| $ c_i $ | The **context word index** used for embedding lookup. |
| $ w_t $ | The **center word** (target word) to be predicted. |
| $ P(w_t \mid \text{context}) $ | The **conditional probability** of $ w_t $ being the correct word, given the context. |
| $ log P(w_t \mid \text{context}) $ | The **log-probability** of predicting $ w_t $ correctly given its context. |


---

#### Parameters Learned

- $ E \in {R}^{V \times d} $: Embedding matrix (input)
- $ E' \in {R}^{V \times d} $: Output matrix (used for prediction)

**Note**: Final embeddings are typically taken from matrix $ E $

---

### Summary

- **Input**: Integer indices of context words  
- **Output**: Predict index of target (center) word  
- **Lookup**: Embed each context word, then average  
- **Train**: With softmax or negative sampling  
- **Goal**: Words that appear in similar contexts should learn similar vector representations

### **Negative Sampling in Word2Vec**

---

#### Why Not Full Softmax?

In the original Skip-gram formulation, the softmax function is used to compute the probability of a context word given a target word:

$$
P(w_o \mid w_t) = \frac{\exp(u_{w_o}^\top v_{w_t})}{\sum_{w=1}^{V} \exp(u_w^\top v_{w_t})}
$$

Where:
- $ v_{w_t} $: input (center) embedding of word $ w_t $
- $ u_{w_o} $: output (context) embedding of word $ w_o $
- $ V $: vocabulary size

**Problem**:  
The denominator sums over **all words in the vocabulary** $ V $ — this is extremely slow for large corpora.

---

#### Solution: Negative Sampling

Instead of updating **all output weights**, **update only a small number of "negative samples"** along with the one positive pair.

---

#### Intuition

- For each training pair $ (w_t, w_o) $:
  - Treat it as a **positive example**
  - Sample $ k $ random words $ \{w_1^{'}, w_2^{'}, ..., w_k^{'}\} $ from the vocabulary — these are **negative samples**
- The goal:
  - **Maximize** the probability of real context word $ w_o $
  - **Minimize** the probability of negative samples $ w_i^{'} $

---

#### Loss Function

For a center word $ w_t $ and a context word $ w_o $, the **negative sampling loss** is:

$$
{L}_{\text{negative-sampling}} = -\log \sigma(u_{w_o}^\top v_{w_t}) - \sum_{i=1}^{k} \log \sigma(-u_{w_i^{'}}^\top v_{w_t})
$$

Where:
- $ \sigma(x) = \frac{1}{1 + \exp(-x)} $  : sigmoid function
- $ u_w $: output embedding  of word $ w $
- $ v_w $: input embedding  of word $ w $
- $ w_o $: actual word (positive sample)
- $ w_i^{'} $: i-th negative sample
- $ k $: number of negative samples

---

#### Training Workflow (Step-by-Step)

1. **Input**: target word $ w_t $, context word $ w_o $
2. **Get embeddings**:
   - $ v_{w_t} \in {R}^d $ from input matrix $ E $
   - $ u_{w_o} \in {R}^d $ from output matrix $ E'$
3. **Sample** $ k $ random words from vocabulary as negative samples
4. **Compute loss** using the formula above
5. **Backpropagate** and update only:
   - $ v_{w_t} $
   - $ u_{w_o} $
   - $ u_{w_i^{'}} $ for all $ i \in [1, k] $

---




#### Why It Works

- Drastically reduces **computation time**.
- Still allows the model to learn **good semantic embeddings**.
- Scales to **billions of tokens** and **millions of words**.
- Enables **mini-batch training** with sparse updates.


---
### **Hierarchical Softmax in Word2Vec**

---

####  Why Another Alternative to Softmax?

Like Negative Sampling, **Hierarchical Softmax (HS)** addresses the computational inefficiency of the full softmax, especially for large vocabularies.

**Key Idea**:  
 Instead of scoring all words in the vocabulary, we build a **binary tree** and compute the probability of a word by traversing the path from the **root to that word's leaf**.

---

#### How Hierarchical Softmax Works

1. Build a **binary tree** where:
   - **Leaves** represent vocabulary words.
   - **Internal nodes** make **binary decisions**: left/right.

2. Each word is uniquely identified by a **path from the root to a leaf**.

3. The model:
   - Learns vector representations for each **internal node**.
   - Predicts the correct **binary decision** at each node in the path.

---

#### Example

Say the word **"sat"** is located along this binary path:
```
Root → Left → Right → Left → [sat]
```

We must predict:
- Step 1: Go **Left**
- Step 2: Go **Right**
- Step 3: Go **Left**

So the model predicts **a sequence of decisions** — **not** the word directly.

---

#### Mathematical Formulation

Let:
- $ w $ : target word to predict
- $ w_t $ : input word
- $ L(w) $ : length of the binary path to $ w $
- $ n_i $ : the $ i^{th} $ internal node along the path to $ w $
- $ s_i \in \{+1, -1\} $: direction at node $ n_i $ (+1 = left, -1 = right)
- $ v_{w_t} $ : input vector for $ w_t $
- $ v_{n_i} $ : vector associated with internal node $ n_i $
- $ \sigma(x) = \frac{1}{1 + \exp(-x)} $

Then:

$$
P(w \mid w_t) = \prod_{i=1}^{L(w)} \sigma\left( s_i \cdot v_{n_i}^\top v_{w_t} \right)
$$

---

#### Loss Function

For predicting word $ w $ given center word $ w_t $, the **loss is**:

$$
{L}_{\text{HS}} = - \sum_{i=1}^{L(w)} \log \sigma\left( s_i \cdot v_{n_i}^\top v_{w_t} \right)
$$

- Similar to Negative Sampling, the model uses **sigmoid** activations at each node.
- Each decision is treated as a **binary classification task**.

---


#### Summary

- Hierarchical Softmax is a tree-based approximation to full softmax.
- Words are predicted by following binary paths.
- Each step in the path is trained via sigmoid + binary cross-entropy.
- Enables fast, full-vocab prediction with logarithmic complexity.

````python
import nltk
from datasets import load_dataset
from tqdm import tqdm
from typing import List
from nltk.tokenize import word_tokenize
from collections import Counter 
import numpy as np
import random


# Download the necessary NLTK tokenizer data
nltk.download("punkt", quiet=True)


EMBEDDING_DIM = 100
MIN_COUNT = 5
WINDOW_SIZE = 2
SEED = 42

def load_hf_data() -> List[List[str]]:
    
    ds = load_dataset("Exploration-Lab/CS779-Fall25", "Assignment-3-word2vec", split="train")
    text_data = ds["text"]
    tokenized_data = [word_tokenize(doc.strip().lower()) 
                      for doc in tqdm(text_data, desc="Tokenizing Sentences") 
                      if doc.strip()]
    
    words = [word for sentence in tokenized_data for word in sentence]
    word_counts = Counter(words)
    pruned_vocab = [(word, count) for word, count in word_counts.most_common() if count >= MIN_COUNT]
    word_to_idx = {word: i for i, (word, count) in enumerate(pruned_vocab)}
    idx_to_word = {i: word for word, i in word_to_idx.items()}
    return tokenized_data, word_to_idx, idx_to_word, word_counts

data, word_to_idx, idx_to_word, word_counts = load_hf_data()

print("Top 10 most common words in vocabulary:")
for i in range(10):
    word = idx_to_word[i]
    print(f"Index {i}: '{word}' (Count: {word_counts[word]})")

VOCAB_SIZE = len(word_to_idx)
print("Final Vocab Size:", VOCAB_SIZE)
````

````python

def build_negative_sampling_table(word_to_idx, word_counts, table_size=1e6):

    # Compute frequencies only for vocab words
    word_freqs = np.array([word_counts[w] for w in word_to_idx.keys()], dtype=np.float64)
    unigram_dist = word_freqs ** 0.75
    unigram_dist /= unigram_dist.sum()

    # Build table for sampling
    table_size = int(table_size)
    counts = np.round(unigram_dist * table_size).astype(int)

    neg_table = []
    for idx, c in enumerate(counts):
        neg_table += [idx] * c
    neg_table = np.array(neg_table)

    return neg_table

neg_table = build_negative_sampling_table(word_to_idx, word_counts)
print("Negative Sampling Table Size:", len(neg_table))
````

````python


def generate_skipgram_pairs(data, word_to_idx, window_size=2):
    pairs = []
    for sentence in data:
        indexed_sentence = [word_to_idx[w] for w in sentence if w in word_to_idx]
        for i, center in enumerate(indexed_sentence):
            window = random.randint(1, window_size) 
            start = max(0, i - window)
            end = min(len(indexed_sentence), i + window + 1)
            for j in range(start, end):
                if j != i:
                    pairs.append((center, indexed_sentence[j]))
    return pairs

pairs = generate_skipgram_pairs(data, word_to_idx, window_size=WINDOW_SIZE)
print("Total training pairs:", len(pairs))
print("Example pair (center, context):", pairs[0])
````

````python
def generate_batches(pairs, batch_size=512):
    random.shuffle(pairs)
    for i in range(0, len(pairs), batch_size):
        yield pairs[i:i+batch_size]


def get_negative_samples_batch(context_batch, neg_table, k):
    
    batch_size = len(context_batch)
    negatives = np.zeros((batch_size, k), dtype=np.int32)
    for i, pos_idx in enumerate(context_batch):
        negs = []
        while len(negs) < k:
            candidate = random.choice(neg_table)
            if candidate != pos_idx:
                negs.append(candidate)
        negatives[i] = negs
    return negatives
````

````python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def train_skipgram_ns(pairs, W_in, W_out, neg_table,
                                lr=0.025, epochs=1, k=5, batch_size=512):
    for epoch in range(epochs):
        total_loss = 0
        batches = list(generate_batches(pairs, batch_size))
        
        for batch in tqdm(batches, desc=f"Epoch {epoch+1}"):
            centers, contexts = zip(*batch)
            centers = np.array(centers)
            contexts = np.array(contexts)

            # Lookup vectors
            v_c = W_in[centers]        
            u_o = W_out[contexts]     

            score_pos = sigmoid(np.sum(v_c * u_o, axis=1)) 
            loss_pos = -np.log(score_pos + 1e-10)

            grad_pos = (1 - score_pos)[:, np.newaxis]       
            W_out[contexts] += lr * grad_pos * v_c
            W_in[centers]   += lr * grad_pos * u_o

            neg_samples = get_negative_samples_batch(contexts, neg_table, k)  
            u_neg = W_out[neg_samples]  
            score_neg = sigmoid(-np.einsum("bd,bkd->bk", v_c, u_neg))  
            loss_neg = -np.log(score_neg + 1e-10)

            grad_neg = (0 - sigmoid(np.einsum("bd,bkd->bk", v_c, u_neg)))[:,:,None] 
            W_out[neg_samples] += lr * grad_neg * v_c[:,None,:]
            W_in[centers]      += lr * np.sum(grad_neg * u_neg, axis=1)

            total_loss += np.sum(loss_pos + np.sum(loss_neg, axis=1))

        avg_loss = total_loss / len(pairs)
        print(f"Epoch {epoch+1} | Avg Loss: {avg_loss:.4f}")
    return W_in
````

````python
np.random.seed(SEED)
W_in = (np.random.rand(VOCAB_SIZE, EMBEDDING_DIM) - 0.5) / EMBEDDING_DIM
W_out = np.zeros((VOCAB_SIZE, EMBEDDING_DIM))  # initialize output embeddings

embeddings = train_skipgram_ns(
    pairs, W_in, W_out, neg_table,
    lr=0.025, epochs=5, k=5, batch_size=512
)
````

````python
import heapq
import numpy as np
import random
from tqdm import tqdm
import pickle
import math


np.random.seed(SEED)
random.seed(SEED)

#Huffman Tree & Paths (Hierarchical Softmax)

def build_huffman_tree_and_paths(itos, word_counts):
    
    # I have kept a structure for nodes: {'freq': , 'word_idx': int or None, 'left': node, 'right': node}
    counter = 0
    heap = []
    for idx, token in enumerate(itos):
        freq = word_counts.get(token, 0)
        node = {'freq': float(freq), 'word_idx': idx, 'left': None, 'right': None, 'index': None}
        heapq.heappush(heap, (node['freq'], counter, node))
        counter += 1

    
    while len(heap) > 1:
        f1, _, n1 = heapq.heappop(heap)
        f2, _, n2 = heapq.heappop(heap)
        parent = {'freq': f1 + f2, 'word_idx': None, 'left': n1, 'right': n2, 'index': None}
        heapq.heappush(heap, (parent['freq'], counter, parent))
        counter += 1
    _, _, root = heapq.heappop(heap)

    idx_counter = 0
    stack = [root]
    while stack:
        node = stack.pop()
        if node is None:
            continue
        if node['word_idx'] is None:
            node['index'] = idx_counter
            idx_counter += 1
            stack.append(node['right'])
            stack.append(node['left'])
    num_internal = idx_counter

    paths = {}
    def dfs(node, path_nodes, path_bits):
        if node is None:
            return
        if node['word_idx'] is not None:
            # leaf
            paths[node['word_idx']] = (list(path_nodes), list(path_bits))
            return
        # left : bit +1
        if node['left'] is not None:
            if node['index'] is None:
                raise RuntimeError("Internal node missing index")
            path_nodes.append(node['index'])
            path_bits.append(1)
            dfs(node['left'], path_nodes, path_bits)
            path_nodes.pop(); path_bits.pop()
        # right : bit -1
        if node['right'] is not None:
            path_nodes.append(node['index'])
            path_bits.append(-1)
            dfs(node['right'], path_nodes, path_bits)
            path_nodes.pop(); path_bits.pop()

    dfs(root, [], [])
    return paths, num_internal

def generate_batches(pairs, batch_size=512):
    """Yield minibatches (list of (center_idx, context_idx)). Shuffles pairs in-place."""
    random.shuffle(pairs)
    for i in range(0, len(pairs), batch_size):
        yield pairs[i:i+batch_size]



import numpy as np
from tqdm import tqdm

def train_skipgram_hs_minibatch(pairs, W_in, paths, num_internal,
                                 epochs=5, batch_size=512, lr=0.025, verbose=True):
    
    V, D = W_in.shape
    rng = np.random.RandomState(SEED)
    E_internal = (rng.randn(num_internal, D).astype(np.float32) * 0.01)

    # Precompute max path length
    max_path_len = max(len(bits) for _, bits in paths.values())

    # Prepare path arrays for all words
    # paths_nodes_array[word_idx] ->  array of node indices
    # paths_bits_array[word_idx] ->  array of bits
    paths_nodes_array = np.full((V, max_path_len), -1, dtype=np.int32)
    paths_bits_array  = np.zeros((V, max_path_len), dtype=np.int32)
    path_lengths = np.zeros(V, dtype=np.int32)

    for word_idx, (nodes, bits) in paths.items():
        path_lengths[word_idx] = len(bits)
        paths_nodes_array[word_idx, :len(bits)] = nodes
        paths_bits_array[word_idx, :len(bits)] = bits

    num_batches = int(np.ceil(len(pairs) / batch_size))

    for epoch in range(1, epochs + 1):
        np.random.shuffle(pairs)
        t_loss = 0.0

        for b in tqdm(range(num_batches), desc=f"HS Epoch {epoch}/{epochs}"):
            batch_pairs = pairs[b*batch_size : (b+1)*batch_size]
            if not batch_pairs:
                continue

            centers = np.array([c for c, t in batch_pairs], dtype=np.int32)
            targets = np.array([t for c, t in batch_pairs], dtype=np.int32)
            batch_size_actual = len(centers)

            # Compute center embeddings
            v_c = W_in[centers] 

            # Fetch path nodes and bits
            max_len = max(path_lengths[targets])
            nodes_batch = paths_nodes_array[targets, :max_len]  
            bits_batch  = paths_bits_array[targets, :max_len]   
            mask = (nodes_batch >= 0).astype(np.float32)      

            # Fetch internal node embeddings for all nodes in batch
            u_n = E_internal[nodes_batch] 

            # Compute dot products: v_c · u_n (broadcast)
            dot = np.einsum('bd,bld->bl', v_c, u_n)  
            s_dot = dot * bits_batch                  # signed dot
            sigma = 1.0 / (1.0 + np.exp(-s_dot))
            t_loss += -np.sum(mask * np.log(np.clip(sigma, 1e-12, 1.0)))

            # Gradients
            g = (sigma - 1.0) * bits_batch           
            g_exp = g[:, :, np.newaxis]             

            grad_u = g_exp * v_c[:, np.newaxis, :] * mask[:, :, np.newaxis] 
            grad_v = np.sum(g_exp * u_n * mask[:, :, np.newaxis], axis=1)  

            # Accumulate gradients into embeddings
            for i in range(batch_size_actual):
                idx = centers[i]
                W_in[idx] -= lr * grad_v[i] / batch_size_actual
                for j, node_idx in enumerate(nodes_batch[i]):
                    if node_idx >= 0:
                        E_internal[node_idx] -= lr * grad_u[i, j] / batch_size_actual

        if verbose:
            avg_loss = t_loss / len(pairs)
            print(f"[HS] Epoch {epoch}/{epochs} finished. Avg loss per example ~ {avg_loss:.6f}")

    return W_in, E_internal

itos_list = [idx_to_word[i] for i in range(len(idx_to_word))]  
paths_map, num_internal_nodes = build_huffman_tree_and_paths(itos_list, word_counts)


print("Num words (vocab):", len(itos_list))
print("Num internal nodes:", num_internal_nodes)
for widx in range(min(5, len(itos_list))):
    nodes, bits = paths_map[widx]
    print("word:", itos_list[widx], "path_len:", len(nodes), "nodes:", nodes[:5], "bits:", bits[:5])

W_in_trained, E_internal_trained = train_skipgram_hs_minibatch(
    pairs=pairs,
    W_in=W_in.copy(),
    paths=paths_map,
    num_internal=num_internal_nodes,
    epochs=3,
    batch_size=512,
    lr=0.025,
    verbose=True
)
````

````python
import numpy as np
import random
from tqdm import tqdm

np.random.seed(42)
random.seed(42)


def generate_cbow_pairs(data, window_size):
    """
    Generate CBOW training pairs: (context_words, target_word)
    Returns list of (context_idx_list, target_idx)
    """
    pairs = []
    for sentence in data:
        sentence_idx = [word_to_idx[w] for w in sentence if w in word_to_idx]
        for i, target in enumerate(sentence_idx):
            context = []
            for j in range(max(0, i - window_size), min(len(sentence_idx), i + window_size + 1)):
                if j != i:
                    context.append(sentence_idx[j])
            if context:
                pairs.append((context, target))
    return pairs


def train_cbow_ns_vectorized(pairs_idx, neg_table, vocab_size, embedding_dim=100,
                             batch_size=512, epochs=5, lr=0.025, neg_samples=5):
    
    W_in = (np.random.rand(vocab_size, embedding_dim) - 0.5) / embedding_dim
    W_out = np.zeros((vocab_size, embedding_dim))
    
    num_pairs = len(pairs_idx)
    print(f"[CBOW + NS Vectorized] Training {num_pairs} pairs, epochs={epochs}, batch_size={batch_size}")
    
    for epoch in range(epochs):
        np.random.shuffle(pairs_idx)
        epoch_loss = 0.0
        
        for batch_start in tqdm(range(0, num_pairs, batch_size), desc=f"Epoch {epoch+1}/{epochs}"):
            batch = pairs_idx[batch_start: batch_start + batch_size]
            context_batch = [c for c, t in batch]
            targets = np.array([t for c, t in batch])
            actual_batch_size = len(targets)
            
            v_c = np.zeros((actual_batch_size, embedding_dim))
            for i, context in enumerate(context_batch):
                v_c[i] = np.mean(W_in[context], axis=0)
            
            
            u_o = W_out[targets]  
            score_pos = sigmoid(np.sum(v_c * u_o, axis=1)) 
            g_pos = score_pos - 1.0  
            epoch_loss += -np.sum(np.log(score_pos + 1e-12))
            
            neg_idxs = get_negative_samples_batch(targets, neg_table, neg_samples)  
            u_k = W_out[neg_idxs]  

            score_neg = sigmoid(np.sum(v_c[:, None, :] * u_k, axis=2)) 

            g_neg = score_neg
            epoch_loss += -np.sum(np.log(1 - score_neg + 1e-12))
            
            #Initialize gradients
            grad_in = np.zeros_like(W_in)
            grad_out = np.zeros_like(W_out)
            
            np.add.at(grad_out, targets, g_pos[:, None] * v_c)
            for i, context in enumerate(context_batch):
                grad_in[context] += g_pos[i] * u_o[i] / len(context)
            
            np.add.at(grad_out, neg_idxs, g_neg[:, :, None] * v_c[:, None, :])
            for i, context in enumerate(context_batch):
                grad_in[context] += np.sum(g_neg[i][:, None] * u_k[i], axis=0) / len(context)
            
            #Gradient step
            W_in -= lr * grad_in / actual_batch_size
            W_out -= lr * grad_out / actual_batch_size
        
        print(f"Epoch {epoch+1} Loss: {epoch_loss/num_pairs:.6f}")
    
    return W_in, W_out

pairs_idx = generate_cbow_pairs(data, WINDOW_SIZE)

neg_table = build_negative_sampling_table(word_to_idx, word_counts, table_size=1_000_000)

W_in_cbow_ns, W_out_cbow_ns = train_cbow_ns_vectorized(
    pairs_idx, neg_table,
    vocab_size=len(word_to_idx),
    embedding_dim=EMBEDDING_DIM,
    batch_size=512,
    epochs=3,
    lr=0.025,
    neg_samples=5
)
````

````python
def train_cbow_hs_vectorized(pairs_idx, itos, word_counts,
                             embedding_dim=100,
                             epochs=5,
                             batch_size=512,
                             lr=0.025,
                             seed=42,
                             verbose=True):
   
    rng = np.random.RandomState(seed)
    V = len(itos)
    D = embedding_dim

    # Build Huffman tree and get paths: word_idx -> (node_idx_list, bits_list)
    paths_map, num_internal = build_huffman_tree_and_paths(itos, word_counts)
   
    print(f"Huffman built: vocab={V}, internal_nodes={num_internal}")

    # Precompute max path length and padded arrays for all words
    max_path_len = max(len(bits) for _, bits in paths_map.values())
    paths_nodes_array = np.full((V, max_path_len), -1, dtype=np.int32)
    paths_bits_array  = np.zeros((V, max_path_len), dtype=np.int32)
    path_lengths = np.zeros(V, dtype=np.int32)
    for widx, (nodes, bits) in paths_map.items():
        L = len(bits)
        path_lengths[widx] = L
        paths_nodes_array[widx, :L] = nodes
        paths_bits_array[widx, :L]  = bits

    # Initialize embeddings
    random_gen = np.random.RandomState(seed)
    W_in = (random_gen.randn(V, D).astype(np.float32) * 0.01)
    E_internal = (random_gen.randn(num_internal, D).astype(np.float32) * 0.01)

    num_pairs = len(pairs_idx)
    num_batches = int(np.ceil(num_pairs / batch_size))
    print(f"[CBOW+HS] Training: {num_pairs} pairs, epochs={epochs}, batch_size={batch_size}, batches={num_batches}")

    for epoch in range(1, epochs + 1):
        random_gen.shuffle(pairs_idx)
        epoch_loss = 0.0

        for b in tqdm(range(num_batches), desc=f"CBOW-HS Epoch {epoch}/{epochs}"):
            batch = pairs_idx[b*batch_size : (b+1)*batch_size]
            if not batch:
                continue

            # unpack contexts and targets
            context_batch = [c for c, t in batch]        # list of lists (variable length)
            targets = np.array([t for c, t in batch], dtype=np.int32)
            B = len(targets)

            # 1) compute averaged context embeddings v_c for the batch (B, D)
            v_c = np.zeros((B, D), dtype=np.float32)
            ctx_lengths = np.empty(B, dtype=np.int32)
            for i, ctx in enumerate(context_batch):
                ctx_lengths[i] = len(ctx)
                v_c[i] = np.mean(W_in[ctx], axis=0)

            # 2) fetch path nodes & bits for targets, restrict to max_len_in_batch to save compute
            max_len_in_batch = int(path_lengths[targets].max())
            nodes_batch = paths_nodes_array[targets, :max_len_in_batch]  # (B, L)
            bits_batch  = paths_bits_array[targets, :max_len_in_batch]   # (B, L)
            mask_batch  = (nodes_batch >= 0).astype(np.float32)          # (B, L)

            # 3) gather internal node embeddings u_n for nodes in batch: (B, L, D)
            u_n = E_internal[nodes_batch]  
            u_n = u_n * mask_batch[:, :, None]

            # 4) compute signed dot products
            dot = np.einsum('bd,bld->bl', v_c, u_n)    # (B, L)
            s_dot = dot * bits_batch                   # (B, L)
            sigma = 1.0 / (1.0 + np.exp(-s_dot))       # (B, L)
            # avoid log(0)
            epoch_loss += -np.sum(mask_batch * np.log(np.clip(sigma, 1e-12, 1.0)))

            # 5) compute gradients
            g = (sigma - 1.0) * bits_batch
            g_masked = g * mask_batch                    # (B, L)
            g_expanded = g_masked[:, :, None]            # (B, L, 1)

            grad_u_batch = g_expanded * v_c[:, None, :]  # (B, L, D)
            grad_v_batch = np.sum(g_expanded * u_n, axis=1)  # (B, D)

            # 6) Accumulate gradients into E_internal (num_internal, D) using np.add.at
            #    flatten nodes and grads for masked positions
            nodes_flat = nodes_batch.reshape(-1)           # (B*L,)
            mask_flat  = mask_batch.reshape(-1)           # (B*L,)
            grad_u_flat = grad_u_batch.reshape(-1, D)     # (B*L, D)
            sel = mask_flat.astype(bool)
            if sel.sum() > 0:
                nodes_sel = nodes_flat[sel].astype(np.int32)
                grad_u_sel = grad_u_flat[sel]
                # accumulate
                np.add.at(E_internal, nodes_sel, - (lr / B) * grad_u_sel)   # update internal nodes

            # 7) Accumulate gradients into W_in for context words:
            ctx_idx_list = []       
            ctx_contribs = []       
            for i, ctx in enumerate(context_batch):
                if len(ctx) == 0:
                    continue
                contrib = - (lr / B) * (grad_v_batch[i] / len(ctx))  
                ctx_idx_list.extend(ctx)
                ctx_contribs.extend([contrib] * len(ctx))

            if ctx_idx_list:
                ctx_idx_arr = np.array(ctx_idx_list, dtype=np.int32)
                ctx_contribs_arr = np.stack(ctx_contribs, axis=0).astype(np.float32)  
                np.add.at(W_in, ctx_idx_arr, ctx_contribs_arr)


        
        avg_loss = epoch_loss / max(1, num_pairs)
        print(f"[CBOW-HS] Epoch {epoch} avg loss per pair: {avg_loss:.6f}")

    return W_in, E_internal, paths_map
````

## **1. Concepts**  
Naive Bayes is a **probabilistic machine learning algorithm** based on **Bayes’ theorem** with the simplifying assumption that all features are conditionally independent given the class. Despite its simplicity, it is widely used in domains such as spam detection, sentiment analysis, and text classification due to its efficiency and effectiveness on high-dimensional data.  

---

## **1.1 Motivation**  

- Many classification tasks involve **large feature spaces** (e.g., words in documents).  
- Complex models struggle with high dimensionality and require significant computation.  
- Naive Bayes offers a **lightweight and interpretable** solution that is both **fast to train and test**.  
- Works surprisingly well even when the independence assumption does not hold perfectly.  

**Example:** Predict whether an email is *spam* or *ham* using word occurrences.  

---

## **1.2 Core Ideas**  

## Naive Bayes Theory

- **Bayes’ Theorem**

$$
P(C \mid X) = \frac{P(X \mid C) \, P(C)}{P(X)}
$$

Where:  
- \($C$\): Class label  
- \($X$\): Feature vector (e.g., tokens in a document)  

---

### **1. Naive Independence Assumption**

$$
P(X \mid C) = \prod_{i=1}^n P(x_i \mid C)
$$


- \($X = (x_1, x_2, \dots, x_n)$\): the feature vector (all attributes/words in a document).  
- \($C$\): the class label (e.g., spam/ham, positive/negative, topic).  
- \($x_i$\): the \($i$\)-th feature of \($X$\) (e.g., presence/absence or count of word \($i$\)).  
- \($n$\): the number of features (e.g., vocabulary size).  
- \($P(X \mid C)$\): probability of observing the full feature vector \($X$\) given class \($C$\).  
- \($P(x_i \mid C)$\): probability of observing the $i$-th feature given class \($C$\).  

👉 The **naive** assumption is that features are conditionally independent given the class, so we can multiply their probabilities.

---

### **2. Classification Rule**

$$
\hat{C} = \arg\max_C \; P(C) \prod_{i=1}^n P(x_i \mid C)
$$

- \($\hat{C}$\): predicted class for the instance (the classifier’s output).  
- \($\arg\max_C$\): choose the class \($C$\) that maximizes the expression.  
- \($P(C)$\): prior probability of class \($C$\) (from class frequencies).  
- \($\prod_{i=1}^n P(x_i \mid C)$\): likelihood of observing \($X$\) under class \($C$\).  

👉 The classifier picks the class that makes the features most probable under Bayes’ rule.


- **Intuition**  
Naive Bayes simplifies classification by assuming features are conditionally independent given the class. Instead of estimating a complex joint probability distribution, it multiplies individual feature likelihoods, making it efficient and effective for high-dimensional data like text.


---

## **1.4 Implementation Details**  

## Naive Bayes Steps

1. **Training Phase**  
   - Compute **prior probabilities** $P(C)$ from class frequencies.  
   - Estimate **conditional probabilities** $P(x_i \mid C)$ from feature counts.  

---

2. **Smoothing**  
- Use **Laplace smoothing** to avoid zero probabilities:  

   $$
   P(x_i \mid C) = \frac{\text{count}(x_i, C) + 1}{\sum_j \text{count}(x_j, C) + |V|}
   $$  
where,
-  $|V|$ = vocabulary size.  
- \($x_i$\): the \($i$\)-th feature of \($X$\) (e.g., presence/absence or count of word \($i$\)).  

---

3. **Prediction Phase**  
- Compute posterior probabilities for each class.  
- Use logarithms to prevent underflow in large feature spaces.  
- Select the class with maximum posterior probability.  


## **1.5 Summary**  

- Naive Bayes is a **probabilistic classifier** grounded in Bayes’ theorem.  
- The “naive” assumption simplifies computation, making it **fast and scalable**.  
- Variants handle different data types: multinomial, Bernoulli, and Gaussian.  
- Despite its simplicity, it is one of the most **robust baseline algorithms** in machine learning, especially in **text classification**.

````python
import numpy as np
import re
import csv
from collections import Counter, defaultdict
from datasets import load_dataset
from sklearn.feature_extraction.text import TfidfVectorizer
````

````python
def train_naive_bayes_tfidf(X, y, vectorizer):

    # Transform documents into TF-IDF matrix
    X_tfidf = vectorizer.fit_transform(X)
    vocab = vectorizer.get_feature_names_out()
    vocab_size = len(vocab)

    total_docs = len(X)
    class_doc_counts = Counter(y)

    # Priors
    prior = {cls: count / total_docs for cls, count in class_doc_counts.items()}

    # Initialize word weights per class
    class_word_weights = {cls: np.ones(vocab_size) for cls in class_doc_counts}  # +1 for Laplace smoothing
    total_class_weights = {cls: vocab_size for cls in class_doc_counts}          # start with |V|

    # Accumulate TF-IDF weights
    for i, label in enumerate(y):
        class_word_weights[label] += X_tfidf[i].toarray().flatten()
        total_class_weights[label] += X_tfidf[i].sum()

    # Likelihoods
    likelihoods = {}
    for cls in class_doc_counts:
        likelihoods[cls] = class_word_weights[cls] / total_class_weights[cls]

    return prior, likelihoods, vocab, vectorizer

def predict_naive_bayes_tfidf(X, prior, likelihoods, vocab, vectorizer):
    X_tfidf = vectorizer.transform(X)
    preds = []

    for i in range(X_tfidf.shape[0]):
        scores = {}
        features = X_tfidf[i].toarray().flatten()
        for cls in prior:
            # log prior
            score = np.log(prior[cls])
            # add log likelihood weighted by tf-idf
            score += (features * np.log(likelihoods[cls])).sum()
            scores[cls] = score
        preds.append(max(scores, key=scores.get))

    return preds
````

````python
import numpy as np
from collections import Counter

def accuracy_score(y_true, y_pred):
    correct = sum(yt == yp for yt, yp in zip(y_true, y_pred))
    return correct / len(y_true)

def precision_recall_f1(y_true, y_pred, average="weighted"):
    labels = set(y_true)
    label_counts = Counter(y_true)
    
    precisions, recalls, f1s = {}, {}, {}
    for label in labels:
        tp = sum((yt == label and yp == label) for yt, yp in zip(y_true, y_pred))
        fp = sum((yt != label and yp == label) for yt, yp in zip(y_true, y_pred))
        fn = sum((yt == label and yp != label) for yt, yp in zip(y_true, y_pred))
        
        precision = tp / (tp + fp) if (tp + fp) > 0 else 0.0
        recall    = tp / (tp + fn) if (tp + fn) > 0 else 0.0
        f1        = (2 * precision * recall / (precision + recall)) if (precision + recall) > 0 else 0.0
        
        precisions[label] = precision
        recalls[label] = recall
        f1s[label] = f1


    total = len(y_true)
    prec = sum(precisions[l] * label_counts[l] for l in labels) / total
    rec  = sum(recalls[l] * label_counts[l] for l in labels) / total
    f1   = sum(f1s[l] * label_counts[l] for l in labels) / total

    return prec, rec, f1
````

# **Task 3 : Expectation Maximization**

## **1. Concepts**

Expectation Maximization (EM) is a **probabilistic optimization algorithm** used to estimate parameters of statistical models when data has **latent (hidden) variables**. It alternates between assigning probabilities to hidden variables (E-step) and maximizing the likelihood of the parameters (M-step). EM is widely applied in clustering (e.g., Gaussian Mixture Models), missing data problems, and probabilistic inference.  

---

## 1.1 Motivation

- Real-world datasets often contain **incomplete or hidden information** (e.g., cluster assignments are unknown).  
- Direct maximization of the likelihood function becomes intractable due to hidden variables.  
- EM provides an **iterative framework** to estimate parameters efficiently by breaking the problem into two simpler steps.  

Example:  
Clustering points into multiple Gaussian distributions when class labels are unknown.  

---

## 1.2 Core Ideas

### (A) Theoretical Explanation

- **Likelihood with hidden variables**:  

$$
P(X \mid \theta) = \sum_Z P(X, Z \mid \theta)
$$  

Where:  
- \( $X$ \): Observed data  
- \( $Z$ \): Latent (hidden) variables  
- \( $\theta$ \): Model parameters  

- **E-step (Expectation)**:  
  Estimate the posterior distribution of hidden variables given current parameters:  

$$
Q(Z) = P(Z \mid X, \theta^{(t)})
$$  

- **M-step (Maximization)**:  
  Update parameters by maximizing the expected log-likelihood:  

$$
\theta^{(t+1)} = \arg\max_\theta \, \mathbb{E}_{Q(Z)}[\log P(X, Z \mid \theta)]
$$  

- **Intuition**:  
  The E-step “fills in” missing/hidden data with probabilities, while the M-step re-estimates parameters as if the hidden data were observed. This alternation improves the likelihood iteratively.  

---

### (B) Example: Gaussian Mixture Model (GMM) Clustering  

Given data points without labels:  

- **E-step**: Compute the probability that each data point belongs to each Gaussian component.  
- **M-step**: Update means, covariances, and mixture weights using these probabilities.  
- Repeat until convergence.  

This way, EM allows clustering without prior labels.  

---

## 1.3 Variants of EM  

1. **Gaussian Mixture Models (GMM-EM)**  
   - Clustering with Gaussian distributions.  

2. **Hidden Markov Models (HMM-EM, i.e., Baum-Welch Algorithm)**  
   - Sequence data with hidden states.  

3. **Soft K-Means**  
   - EM with isotropic Gaussians and uniform variance assumption.  

---

## 1.4 Implementation Details  

1. **Initialization**  
   - Start with random guesses for parameters (e.g., means and covariances in GMM).  

2. **E-step**  
   - Calculate the probability distribution of hidden variables given observed data.  

3. **M-step**  
   - Update parameter estimates by maximizing the expected log-likelihood.  

4. **Convergence**  
   - Stop when parameters stabilize or the log-likelihood improvement falls below a threshold.   

---

## 1.5 Summary  

- Expectation Maximization is an **iterative optimization algorithm** for parameter estimation in models with hidden variables.  
- Alternates between **E-step (inferring hidden variables)** and **M-step (optimizing parameters)**.  
- Widely applied in **clustering, HMMs, and incomplete-data problems**.  
- Sensitive to initialization but remains a cornerstone in unsupervised learning.

````python
import time
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from datasets import load_dataset

#This function helps in preventing overflows if the 
def row_logsumexp(a):
    maxi = np.max(a, axis=1)

    smaller = a - maxi[:, np.newaxis]
    s = np.sum(np.exp(smaller), axis=1)
    
    return maxi.reshape(-1, 1), maxi.reshape(-1, 1) + np.log(s.reshape(-1, 1) + 1e-300)

def initialize_params(X, K, alpha, random_state=42):
    random_generator = np.random.RandomState(random_state)
    N, V = X.shape

    pi = np.array([1.0 / K] * K)

    #randomly selecting K words to initialise the clusters
    starters = random_generator.choice(N, size=K)
    phi = np.zeros((K, V))

    for j, idx in enumerate(starters):
        row = X[idx]
        phi[j, :] = np.asarray(row.todense()).ravel()
    
    
    phi = phi + alpha  #Laplace Smoothing 
    phi = phi / (phi.sum(axis=1, keepdims=True) + 1e-300)
    return pi, phi

def em_fit(X, K=10, alpha=1e-2, max_iter=100, tol=1e-4, verbose=True, random_state=42):
    
    N, V = X.shape
    pi, phi = initialize_params(X, K, alpha, random_state)
    prev_loglik = None
    history = []

    for it in range(1, max_iter + 1):
        t0 = time.time()
        log_phi = np.log(phi + 1e-300)
        log_pi = np.log(pi + 1e-300)

        WX = X.dot(log_phi.T)
        log_unnormalized = WX + log_pi[np.newaxis, :]
        _, logsum = row_logsumexp(log_unnormalized)
        log_resp = log_unnormalized - logsum
        resp = np.exp(log_resp)

        loglik = np.sum(logsum)
        history.append(loglik)

        r_sum = resp.sum(axis=0)
        pi = (r_sum / N).astype(float)

        phi_numer_VxK = X.T.dot(resp)
        phi_numer = phi_numer_VxK.T
        phi = (phi_numer + alpha)
        phi /= (phi.sum(axis=1, keepdims=True) + 1e-300)

        dt = time.time() - t0
        if verbose:
            print(f"Iter {it:3d}: loglik={loglik:.6f}, time={dt:.2f}s")

        if prev_loglik is not None:
            improvement = loglik - prev_loglik
            if abs(improvement) < tol * max(1.0, abs(loglik)):
                if verbose:
                    print(f"Converged at iter={it}")
                break
        prev_loglik = loglik

    return pi, phi, history

def em_predict(X, pi, phi):
    log_phi = np.log(phi + 1e-300)
    log_pi = np.log(pi + 1e-300)
    
    WX = X.dot(log_phi.T)

    log_scores = WX + log_pi 
    labels = np.argmax(log_scores, axis=1)
    return labels



n_clusters = 10
max_features = 100000
min_df = 2
token_pattern = r"(?u)\b\w\w+\b"

alpha = 1e-2
max_iter = 100
tol = 1e-4

random_seed = 42

print("Loading dataset...")

ds_train = load_dataset("Exploration-Lab/CS779-Fall25", "Assignment-3-em", split="train")
ds_test = load_dataset("Exploration-Lab/CS779-Fall25", "Assignment-3-em", split="test")

text_col = ['text']
train_texts = [ex['text'] for ex in ds_train]
test_texts = [ex['text'] for ex in ds_test]

vectorizer = CountVectorizer(max_features=max_features, min_df=min_df, token_pattern=token_pattern)
X_train = vectorizer.fit_transform(train_texts)
X_test = vectorizer.transform(test_texts)
print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")

pi, phi, history = em_fit(X_train, K=n_clusters, alpha=alpha,
                          max_iter=max_iter, tol=tol,
                          verbose=True, random_state=random_seed)

labels = em_predict(X_test, pi, phi)
np.savetxt("em_predictions.csv", labels.astype(int), fmt='%d')
print(f"Saved predictions to em_predictions.csv")
````

Q.1 Why is Expectation Maximization considered an iterative optimization algorithm, and how does it differ from direct likelihood maximization?

The Expectation-Maximization (EM) algorithm helps in estimating model parameters that involve unobserved variables. The algorithm operates in cycles consisting of an Expectation step, where the algorithm estimates missing data based on the current parameters, and a Maximization step that uses this "completed data" to adjust the parameters. This process continues iteratively until the estimates stabilize. Hence, EM is considered an iterative optimization algorithm

EM fundamentally differs from direct likelihood maximization, which evaluates the parameters in a single step. The direct approach typically fails when latent variables exist, as missing data renders the likelihood function impossible to optimize. 

Q.2 What is the role of the E-step and M-step in EM, and why are they repeated until convergence?  

 The E-step (Expectation) acts as a guess, using the current model parameters to calculate the most probable values for the hidden or latent variables. Following this, the M-step (Maximization) updates the model's parameters to best fit the data that has now been "completed" by the E-step's probabilistic guesses. This two-step cycle is repeated because each iteration is guaranteed to improve the model's likelihood, essentially climbing a hill towards the best solution. The process continues until convergence, the point where the parameter estimates stabilize and further iterations yield no significant improvement, ensuring that a final, optimal solution (a local maximum) has been reached.

Q.3 Why is EM sensitive to initialization, and how can this problem be mitigated in practice?  

The Expectation-Maximization (EM) algorithm's sensitivity to initialization stems from its method of optimizing a likelihood function that may have multiple peaks, or local maxima. EM starts at an initial point (the initial parameter guess) and iteratively moves uphill until it reaches a peak. However, it cannot distinguish between a local maximum and the global maximum. Consequently, a poor starting point can lead it to converge to a suboptimal solution. To mitigate this, a common strategy is to run the algorithm multiple times with different random initializations and then select the run that yields the highest final likelihood. Another approach is to use an informed method, such as the output of a simpler, non-iterative algorithm (like K-Means for Gaussian Mixture Models), to provide a more sensible starting point, increasing the chances of finding the global maximum.

Q.4 What are some advantages and disadvantages of EM compared to other clustering/optimization algorithms (e.g., K-Means, gradient-based methods)?  

The main advantage of the Expectation-Maximization (EM) algorithm, especially in clustering, is its flexibility and solid statistical basis. Unlike K-Means, which gives each data point a single cluster and assumes spherical clusters, EM provides a probability for all clusters for every point. It calculates the probability of a point belonging to each cluster. This feature allows it to handle complex, overlapping, and non-spherical clusters effectively, as shown in Gaussian Mixture Models. 

However, EM has some drawbacks. It is sensitive to the initial parameters and can get stuck in unoptimal solutions. It can also be computationally expensive and may take time to converge, especially with large datasets. The E-step requires calculating probabilities for every data point in relation to each cluster. In contrast, K-Means is much quicker and easier to implement. Gradient-based methods like EM provide a more general optimization approach that isn't limited to a latent variable structure, but they can be more challenging to adjust.
